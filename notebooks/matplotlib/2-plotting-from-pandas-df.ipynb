# Plotting desde DataFrames de Pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["B", "C"]).cumsum()

df["A"] = pd.Series(list(range(len(df))))

df.plot(x="A", y="B");
```

```python
data = pd.read_csv("../../data/raw/car-sales.csv")
car_sales = pd.DataFrame(data)
car_sales
```

```python
# Limpiar la columna "Price"
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2024",  periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum() # Convierte los valores a tipo entero y aplica la suma acumulativa
car_sales
```

¿Cuándo usar `cumsum()`?
* Seguimiento acumulativo de ventas, ingresos o cualquier métrica que aumente progresivamente.
* Cálculo de totales acumulativos en series temporales.
* Visualización de progresos o tendencias en datos acumulados.

Este método es particularmente útil en contextos financieros, ventas, inventarios y análisis temporal.

## Plotting con pandas usando el método OO

**¿Cuándo usar `pyplot` vs el método OO de matplotlib (enfoque orientado a objetos)?**

Recomendación General:
* Usa pyplot para tareas rápidas y exploratorias.
* Elige el método OO para gráficos avanzados o cuando necesites control total sobre los elementos del gráfico.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Método pyplot
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# Método OO
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
```

```python
# Método OO
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

ax.set_xlim([45, 100])
```

```python
# Método OO
fix, ax = plt.subplots(figsize=(10,6))

# Gráfico de dispersión
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Personalizar el gráfico
ax.set(title="Enfermedad cardiaca y Niveles de Colesterol",
       xlabel="Edad",
       ylabel="Colesterol")

# Añadir leyenda
ax.legend(*scatter.legend_elements(), title="Target");

# Añadir línea horizontal que representa la media
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
over_50.head()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8,8));

scatter = axs[0].scatter(x=over_50["age"],
                         y=over_50["chol"],
                         c=over_50["target"])

axs[0].set(title="Enfermedad cardiaca y Niveles de Colesterol",
       xlabel="Edad",
       ylabel="Colesterol")

axs[0].legend(*scatter.legend_elements(), title="Target")
axs[0].axhline(over_50["chol"].mean(), linestyle="--");

scatter = axs[1].scatter(x=over_50["age"],
                         y=over_50["thalach"], # Frecuencia cardiaca máxima alcanzada
                         c=over_50["target"])

axs[1].set(title="Enfermedad cardiaca y Frecuencia cardiaca máxima",
       xlabel="Edad",
       ylabel="FC max")

axs[1].legend(*scatter.legend_elements(), title="Target")
axs[1].axhline(over_50["thalach"].mean(), linestyle="--");

# Añadir título a la figura
fig.suptitle("Análisis Enfermedad Cardiaca", fontsize=16, fontweight="bold")
fig.tight_layout() # Ajusta automáticamente el espaciado entre los subplots para evitar superposiciones
```
