### Formas de crear gráficos

```python
# Primer método (confuso)
fig = plt.figure() # crea una figura
ax = fig.add_subplot() # añade axes
plt.show()
```

```python
# Segundo método (confuso)
fig = plt.figure() # crea una figura
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # añade axes
plt.show()
```

```python
# Tercer método (recomendado) (más fácil y robusto)
fig, ax = plt.subplots() 
ax.plot(x, y) # añade axes
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]) # añade datos
```

### Ejemplo de flujo de trabajo en Matplotlib

```python
# 0. Importar matplotlib
import matplotlib.pyplot as plt

# 1. Preparar los datos
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Personalizar el gráfico
ax.set(title="Gráfico simple",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Guardar y mostrar
fig.savefig("../../assets/section-6/sample-plot.png")
```

### Haciendo figuras con NumPy arrays

Queremos hacer:
* Line plot (gráfico de línea)
* Scatter plot (diagrama de dispersión)
* Bar plot (gráfico de barras)
* Histogram (histograma)
* Subplot (múltiples gráficos dentro de una misma figura)

```python
# Crear algunos datos
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Representar los datos y crear una linea en el gráfico
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Usamos los mismos datos para hacer un diagrama de dispersión (scatter)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Otro diagrama de dispersión
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Crear un gráfico desde un diccionario
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),
       nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Precio (€)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Dos opciones para subplots

```python
# Opción 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
# Gráfico para cada eje
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Opción 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Gráfico para cada eje
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
