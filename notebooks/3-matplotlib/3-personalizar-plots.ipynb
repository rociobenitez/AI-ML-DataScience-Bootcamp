```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../data/raw/heart-disease.csv")
data = pd.DataFrame(data)
data.head()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8,8))

# Primer gráfico
scatter = axs[0].scatter(x=data["age"],
                         y=data["chol"],
                         c=data["target"])

axs[0].set(title="Enfermedad cardiaca y Niveles de colesterol",
           xlabel="Edad",
           ylabel="Colesterol")

axs[0].legend(*scatter.legend_elements(), title="Target")
axs[0].axhline(data["chol"].mean(), linestyle="--")

# Segundo gráfico
scatter = axs[1].scatter(x=data["age"],
                         y=data["thalach"],
                         c=data["target"])

axs[1].set(title="Enfermedad cardiaca y Frecuencia cardiaca máxima",
           xlabel="Edad",
           ylabel="FC max")

axs[1].legend(*scatter.legend_elements(), title="Target")
axs[1].axhline(data["thalach"].mean(), linestyle="--")

# Añadir título a la figura
fig.suptitle("Análisis Enfermedad Cardiaca", fontsize=16, fontweight="bold")
fig.tight_layout()
```

```python
plt.style.use('seaborn-v0_8-dark-palette')

fig, ax = plt.subplots(figsize=(10, 8))

# Crear scatter plot categorizado por el target
scatter = ax.scatter(x=data["age"], 
                     y=data["chol"], 
                     c=data["target"], 
                     cmap="coolwarm", 
                     label=["Sin enfermedad", "Con enfermedad"])

# Etiquetas y título
ax.set(title="Relación entre Edad y Colesterol según Enfermedad",
       xlabel="Edad",
       ylabel="Colesterol")

ax.axhline(y=data["chol"].mean(), linestyle="--")

# Crear la leyenda
legend = ax.legend(*scatter.legend_elements(), title="Estado de Enfermedad")
ax.add_artist(legend)
plt.show()
```

### Personalizar gráficos

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8,8))

# Primer gráfico
scatter = axs[0].scatter(x=data["age"],
                         y=data["chol"],
                         c=data["target"],
                         cmap="coolwarm")


axs[0].set(title="Enfermedad cardiaca y Niveles de colesterol",
           xlabel="Edad",
           ylabel="Colesterol")

axs[0].title.set_fontsize(14)  # Cambiar el tamaño del título
axs[0].set_xlim([50,80])       # Ajustar el inicio del eje x (mayores de 50 años)
axs[0].legend(*scatter.legend_elements(), title="Enfermedad")
axs[0].axhline(data["chol"].mean(), linestyle="--")

# Segundo gráfico
scatter = axs[1].scatter(x=data["age"],
                         y=data["thalach"],
                         c=data["target"],
                         cmap="coolwarm")

axs[1].set(title="Enfermedad cardiaca y Frecuencia cardiaca máxima",
           xlabel="Edad",
           ylabel="FC max")

axs[1].title.set_fontsize(14)  # Cambiar el tamaño del título
axs[1].set_xlim([50,80])       # Ajustar el inicio del eje x (mayores de 50 años)
axs[1].set_ylim([60,200])      # Limitar el rango del eje y (de 60 a 200 FC max)
axs[1].legend(*scatter.legend_elements(), title="Enfermedad")
axs[1].axhline(data["thalach"].mean(), linestyle="--")

# Añadir título a la figura
fig.suptitle("Análisis Enfermedad Cardiaca", fontsize=16, fontweight="bold")

# Ajustar espacios entre gráficos
fig.tight_layout()
```
