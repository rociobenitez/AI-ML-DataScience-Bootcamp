## Ejemplo de flujo de trabajo en Matplotlib

```python
# 0. Importar matplotlib
import matplotlib.pyplot as plt

# 1. Preparar los datos
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Personalizar el gr√°fico
ax.set(title="Gr√°fico simple",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Guardar y mostrar
fig.savefig("../../assets/section-6/sample-plot.png")
```

## Tipos de gr√°ficos m√°s comunes utilizando NumPy arrays

- Para saber qu√© tipo de gr√°fico utilizar debemos familiarizarnos con los datos y luego ver cu√°l se adapta mejor a ellos.
- Las visualizaciones de Matplotlib se construyen a partir de arrays de NumPy.

* `Line plot` (gr√°fico de l√≠nea)
* `Scatter plot` (diagrama de dispersi√≥n)
* `Bar plot` (gr√°fico de barras)
* `Histogram` (histograma)
* `Subplot` (m√∫ltiples gr√°ficos dentro de una misma figura)

### Line (L√≠nea)

```python
# Crear algunos datos
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Representar los datos y crear una linea en el gr√°fico
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

### Scatter (diagrama de dispersi√≥n)

```python
# Usamos los mismos datos para hacer un diagrama de dispersi√≥n (scatter)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Otro diagrama de dispersi√≥n
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar (gr√°fico de barras)

* Horizontal
* Vertical

```python
# Crear un gr√°fico desde un diccionario
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),
       nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Precio (‚Ç¨)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

### Histogram (hist)

```python
x = np.random.randn(1000) # distribuci√≥n normal
fig, ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000) # distribuci√≥n aleatoria
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplots

* M√∫ltiples gr√°ficos en una sola figura

```python
# Opci√≥n 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
# Gr√°fico para cada eje
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Opci√≥n 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Gr√°fico para cada eje
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
