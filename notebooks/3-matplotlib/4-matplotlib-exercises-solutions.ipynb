Primero necesitaremos importar pandas y crear un DataFrame con el que trabajar.

```python
# Import pandas as pd
import pandas as pd
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
data = pd.read_csv("../../data/raw/car-sales.csv")
car_sales = pd.DataFrame(data)
car_sales.head()
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales["Price"].plot()
```

쯇or qu칠 no funciona?

Consejo: No son datos num칠ricos.

En el proceso de convertirlo a datos num칠ricos, creemos otra columna que sume el total de ventas y otra que muestre en qu칠 fecha se vendi칩 el coche.

Consejo: Para agregar una columna de manera acumulativa, busca la funci칩n cumsum(). Y para crear una columna de fechas, busca la funci칩n date_range().

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales["Price"] = car_sales["Price"].str.replace("[\.\,\$]", '', regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales.head()
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Sale Date"] = pd.date_range("12/6/2024", periods=len(car_sales))

# View the car_sales DataFrame
car_sales.head()
```

Ahora que tenemos una columna num칠rica (`Total Sales`) y una columna de fechas (`Sale Date`), vamos a visualizarlas.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x="Sale Date", y="Total Sales");
```

Ahora intentemos comparar dos variables con respecto a la variable objetivo.

M치s espec칤ficamente, veremos c칩mo la edad y el colesterol combinados afectan la variable objetivo en **pacientes mayores de 50 a침os**.

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease["age"] > 50]


# Create the plot
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Enfermedad cardiaca y Niveles de colesterol",
       xlabel="Edad",
       ylabel="Colesterol")

ax.legend(*scatter.legend_elements(), title="Enfermedad")

# Add a meanline
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

Hermoso, ahora has creado un gr치fico de dos variables diferentes, cambiemos el estilo.

```python
# Check what styles are available under plt
plt.style.available
```

```python
# Change the style to use "seaborn-whitegrid"
plt.style.use("seaborn-v0_8-whitegrid")
```

Ahora que se ha cambiado el estilo, volvamos a graficar la misma figura de arriba y veamos c칩mo se ve.

```python
# Reproduce the same figure as above with the "seaborn-whitegrid" style

# Create the plot
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Enfermedad cardiaca y Niveles de colesterol",
       xlabel="Edad",
       ylabel="Colesterol")

ax.legend(*scatter.legend_elements(), title="Enfermedad")

# Add a meanline
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

Maravilloso, has cambiado el estilo de los gr치ficos y la figura se ve diferente, pero los puntos no son de un muy buen color.

Cambiemos el par치metro `cmap` de `scatter()` as칤 como el par치metro `color` de `axhline()` para solucionarlo.

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter')

# Customize the plot
ax.set(title="Enfermedad cardiaca y Niveles de colesterol",
       xlabel="Edad",
       ylabel="Colesterol")

ax.legend(*scatter.legend_elements(), title="Enfermedad")

# Add a meanline
ax.axhline(over_50["chol"].mean(), linestyle="--", color="red");
```
