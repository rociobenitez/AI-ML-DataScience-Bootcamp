```python
# Importaciones est√°ndar
import pandas as pd
import numpy as np
import sklearn
```

```python
sklearn.show_versions() # Para ver la versi√≥n de Scikit-Learn
```

### 1. Obtener y preparar los datos

```python
data = pd.read_csv("../../data/raw/heart-disease.csv")
heart_disease = pd.DataFrame(data)
heart_disease.head()
```

Con este ejemplo, vamos a usar todas las columnas excepto la columna objetivo (target) para predecir la columna objetivo.

En otras palabras, utilizar los datos m√©dicos y demogr√°ficos de un paciente para predecir si tiene o no enfermedad card√≠aca.

### 2. Dividir los datos

```python
# Crear X (matriz de caracter√≠sticas)
X = heart_disease.drop("target", axis=1) # Todas las columnas menos la variable objetivo (target)

# Crear y (etiquetas)
y = heart_disease["target"] # La columna/variable objetivo (target)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Ver las dimensiones de los datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 3. Elegir el mejor modelo y sus hiperpar√°metros

Puedes usar el [mapa de machine learning de Scikit-Learn](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html).

<img src="../../assets/section-7/ml_map.svg" width=700/>

En Scikit-Learn, los modelos de aprendizaje autom√°tico se conocen como **estimadores.**

En este caso, dado que estamos trabajando en un problema de **clasificaci√≥n**, hemos elegido el estimador `RandomForestClassifier`, que es parte del m√≥dulo de ensamblajes.

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# Hiperpar√°metros por defecto
clf.get_params()
```

### 4. Ajustar el modelo a los datos de entrenamiento

Un modelo (intentar√°) aprender los patrones en un conjunto de datos al llamar a la funci√≥n `fit()` y pasarle los datos.

```python
clf.fit(X_train, y_train)
```

```python
X_train
```

### 5. Hacer predicciones

Una vez que un modelo ha aprendido patrones en los datos, puedes usarlos para hacer una predicci√≥n con la funci√≥n `predict()`.

```python
# Hacer predicciones
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

### 6. Evaluar el modelo con los datos de entrenamiento y test

Un modelo/estimador entrenado se puede evaluar llamando a la funci√≥n score() y pas√°ndole una colecci√≥n de datos.

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
    
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

### 7. Mejorar el modelo (hyperparameter tuning)

```python
# Probar diferentes valores de n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Probando el modelo con {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Accuracy del modelo en conjunto de prueba: {clf.score(X_test, y_test) * 100:.2f}%") # Porcentaje y limitar decimales
    print("")
```

**Nota:** Es una buena pr√°ctica probar diferentes hiperp√°rametros con un conjunto de validaci√≥n o validaci√≥n cruzada.

```python
from sklearn.model_selection import cross_val_score

# Probar diferentes n√∫meros de estimadores con cross-validation y sin cross-validation
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accruacy on test set: {model.score(X_test, y_test)}")
    print(f"Cross-validation score: {np.mean(cross_val_score(model, X, y, cv=5)) * 100}%")
    print("")
```

### 8. Guardar y cargar el modelo

- Un modelo entrenado se puede exportar y guardar para que pueda ser importado y utilizado m√°s tarde.
- Una forma de guardar un modelo es usar el m√≥dulo `pickle` de Python.

```python
# Guardar el modelo
import pickle

pickle.dump(clf, open("../../models/random_forest_model_1.pkl", "wb"))
```

```python
# Cargar el modelo
loaded_model = pickle.load(open("../../models/random_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
